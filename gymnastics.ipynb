```python
from bs4 import BeautifulSoup # this module helps in web scrapping.
import requests  # this module helps us to download a web page
print("success!")
```

```python
URL = "https://thegymter.net/the-best-scores-in-2021/"
page = requests.get(URL).text
soup = BeautifulSoup(page, "html5lib")
```

```python
tables = soup.find_all('table')
len(tables)
```

```python
for index,table in enumerate(tables):
    table_index = index
    print(tables[table_index])
```

```python
import pandas as pd
```

```python
gym_data = pd.DataFrame(columns=["Rank", "Athlete", "Nation", "Meet", "Score"])

for tr in soup.find_all("tr"):
    tds = tr.find_all('td')
    rank = tds[0].text
    athlete = tds[1].text
    nation = tds[2].text
    meet = tds[3].text
    score = tds[4].text
    gym_data = gym_data.append({"Rank":rank, "Athlete":athlete, "Nation":nation, "Meet":meet, "Score":score}, ignore_index=True)
gym_data
```
